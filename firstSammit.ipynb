## 2. Theoretical Background
### 2.1 Linear Algebra Concepts:
1. **Scalar** : Scalar is a single numerical value that conveys magnitude but no direction or dimension. It is a zero-dimensional entity that acts as a fundamental unit of computation. Scalars are 0 th ordered Tensors.
2. **Vector**: Vector is a single dimensional array that has both magnitude and direction or dimension. It is a 1st-order Tensor. Vectors represent features, weights or data points where each dimension reflects one characteristic or attribute.
3. **Matrices**: Matrices are rectangular arrays or two-dimensional grid consisting of numbers and are an example of 2nd-order tensors.A matrix is a two-dimensional grid or rectangular array of numbers arranged in rows and columns. THe rows and columns denote obseravtions and features resepectively. They are used to represent datasets, transformation functions and neural network weights.
4. **Tensors**: A tensor is a multi-dimensional generalization of scalars, vectors and matrices. It can represent data in three or more dimensions making it suitable for modeling highly complex relationships across multiple axes. In deep learning, tensors are used to store and process data such as images, videos or sequences where each dimension captures a specific feature or time-step.


| **Aspect** | **Scalar**| **Vector** | **Matrix** | **Tensor** |
| :--- | :--- | :--- | :--- | :--- |
| **Dimensionality** | 0D | 1D | 2D | ≥ 3D |
| **Representation** | Single numerical value | Ordered array of values | Two-dimensional grid of values | Multi-dimensional array of values |
| **Usage** | Represent basic quantities | Represent features or directions | Organize data in tabular form | Represent complex data structures |
| **Examples** | Error metric, probability | Feature vector, gradient | Data matrix, weight matrix | Image tensor, video tensor |

## 2.2 Probability Concepts:
1. Probability Distribution: A probability distribution is a mathematical function that describes the probability of different possible values of a variable. Probability distributions are often depicted using graphs or probability tables.<br>
Types of distributions:<br>
i. Uniform Distribution: All outcomes are equally likely.<br>
ii. Normal (Gaussian) Distribution: Symmetric, bell-shaped curve, defined by mean (μ) and standard deviation (σ).<br>
iii. Bernoulli Distribution: Models binary outcomes (e.g., coin flip).<br>

2. Entropy: In Machine Learning, entropy measures the level of disorder or uncertainty in a given dataset or system. It is a metric that quantifies the amount of information in a dataset, and it is commonly used to evaluate the quality of a model and its ability to make accurate predictions. In information theory,entropy quantifies uncertainty and variability in data. High entropy indicates high uncertainty in the system, while low entropy indicates low uncertainty.

3. Bayes' Theorem: Bayes' Theorem is a mathematical formula that determines conditional probability, which is the chance that something happens based on what has already happened in similar situations. It can help revise or update an existing prediction or theory given new evidence.

## 2.3 Tensor Operations 
### 2.3.1. Element Wise Operations:
- Addition and Subtraction: Performed on tensors of same shape where the corresponding elements can be added or subtracted.
- Hadamard Product: Element wise matrix multiplication that is different from normal matrix multiplication. Both the tenors need to be of same size.
- Activation Functions:  Performing a mathematical operation on every individual element within a tensor. Example : ReLU and Sigmoid functions.

### 2.3.2. Linear Algebra Operations:
- Dot Product / Matrix Multiplication: The most common operation in neural networks, computed by multiplying matching elements and summing them, that is used to compute the output of a dense (fully connected) layer by multiplying input features by a weight matrix.
- Transpose: Swapping the axes of a tensor (e.g., flipping a matrix's rows and columns). Shuffling the dimensions of a tensor, such as changing a (3X4X5) tensor to (5X4X3) tensor by swapping axes.
- Tensor Contraction: A generalization of the dot product to higher dimensions, where products are summed over specific axes. This reduces the rank (order) of a tensor by summing over repeated indices.
- Eigen Vectors and Eigen Values : These represent invariant directions and scaling factors, respectively, where the tensor transforms a vector without changing its direction. Eigenvalues are scalars λ that satisfy the equation \(Av = λv\), where \(A\) is a square matrix and \(v\) is a non-zero vector (the eigenvector).Eigenvectors are vectors that, when multiplied by the matrix, result in a scaled version of themselves

### 2.3.3. Shape Manipulations
- Reshaping: Altering the dimensions without altering the underlying data. Example: Convert a (3X4) tensor into (6X2) tensor , while retaining the total number of elements.
- Broadcasting: Automatically "stretching" a smaller tensor (like a scalar or vector) to match the shape of a larger tensor during an operation, allowing implicit element-wise operations between arrays of different shapes.
- Flattening: Converting a multi-dimensional tensor (like a 2D image) into a 1D vector, typically before entering a final classification layer. 

### 2.3.4. Reduction Operations
- Sum/Mean: Calculating the average or total across a specific axis (e.g., averaging pixel values).
- Argmax: Finding the index of the maximum value, often used to determine the final predicted class in a classification model.

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Define sample tensors
A_np = np.random.rand(3, 3)
B_np = np.random.rand(3, 3)
A_tf = tf.constant(A_np, dtype=tf.float32)
B_tf = tf.constant(B_np, dtype=tf.float32)

# Create a figure with a grid layout (14 rows, 2 columns)
fig = plt.figure(figsize=(25, 35))
gs = GridSpec(14, 2, figure=fig)  # Corrected: 14 rows

# Helper function to plot tensors with titles and colorbars
def plot_tensor(ax, title, tensor, lib="NumPy"):
    im = ax.imshow(tensor, cmap='viridis', aspect='auto')
    ax.set_title(f"{title} ({lib})", fontsize=10, pad=10)
    ax.set_xticks([])
    ax.set_yticks([])
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

# --- 2.3.1 Element-wise Operations ---
# Addition
C_np = A_np + B_np
C_tf = A_tf + B_tf
plot_tensor(fig.add_subplot(gs[0, 0]), "Addition", C_np, "NumPy")
plot_tensor(fig.add_subplot(gs[0, 1]), "Addition", C_tf.numpy(), "TensorFlow")

# Subtraction
D_np = A_np - B_np
D_tf = A_tf - B_tf
plot_tensor(fig.add_subplot(gs[1, 0]), "Subtraction", D_np, "NumPy")
plot_tensor(fig.add_subplot(gs[1, 1]), "Subtraction", D_tf.numpy(), "TensorFlow")

# Hadamard Product
E_np = A_np * B_np
E_tf = A_tf * B_tf
plot_tensor(fig.add_subplot(gs[2, 0]), "Hadamard Product", E_np, "NumPy")
plot_tensor(fig.add_subplot(gs[2, 1]), "Hadamard Product", E_tf.numpy(), "TensorFlow")

# ReLU
F_np = np.maximum(0, A_np)
F_tf = tf.nn.relu(A_tf)
plot_tensor(fig.add_subplot(gs[3, 0]), "ReLU", F_np, "NumPy")
plot_tensor(fig.add_subplot(gs[3, 1]), "ReLU", F_tf.numpy(), "TensorFlow")

# Sigmoid
G_np = 1 / (1 + np.exp(-A_np))
G_tf = tf.nn.sigmoid(A_tf)
plot_tensor(fig.add_subplot(gs[4, 0]), "Sigmoid", G_np, "NumPy")
plot_tensor(fig.add_subplot(gs[4, 1]), "Sigmoid", G_tf.numpy(), "TensorFlow")

# --- 2.3.2 Linear Algebra Operations ---
# Matrix Multiplication
H_np = A_np @ B_np
H_tf = tf.matmul(A_tf, B_tf)
plot_tensor(fig.add_subplot(gs[5, 0]), "Matrix Multiplication", H_np, "NumPy")
plot_tensor(fig.add_subplot(gs[5, 1]), "Matrix Multiplication", H_tf.numpy(), "TensorFlow")

# Transpose
I_np = A_np.T
I_tf = tf.transpose(A_tf)
plot_tensor(fig.add_subplot(gs[6, 0]), "Transpose", I_np, "NumPy")
plot_tensor(fig.add_subplot(gs[6, 1]), "Transpose", I_tf.numpy(), "TensorFlow")

# Eigenvalues and Eigenvectors
eigvals_np, eigvecs_np = np.linalg.eig(A_np)
eigvals_tf, eigvecs_tf = tf.linalg.eigh(A_tf)
ax_eig_np = fig.add_subplot(gs[7, 0])
ax_eig_np.bar(range(len(eigvals_np)), eigvals_np, color='blue')
ax_eig_np.set_title("Eigenvalues (NumPy)", fontsize=10, pad=10)
ax_eig_tf = fig.add_subplot(gs[7, 1])
ax_eig_tf.bar(range(len(eigvals_tf)), eigvals_tf.numpy(), color='orange')
ax_eig_tf.set_title("Eigenvalues (TensorFlow)", fontsize=10, pad=10)

# --- 2.3.3 Shape Manipulations ---
# Reshaping
J_np = A_np.reshape(9, 1)
J_tf = tf.reshape(A_tf, (9, 1))
plot_tensor(fig.add_subplot(gs[8, 0]), "Reshape", J_np, "NumPy")
plot_tensor(fig.add_subplot(gs[8, 1]), "Reshape", J_tf.numpy(), "TensorFlow")

# Broadcasting
K_np = A_np + np.array([1, 2, 3])
K_tf = A_tf + tf.constant([1, 2, 3], dtype=tf.float32)
plot_tensor(fig.add_subplot(gs[9, 0]), "Broadcasting", K_np, "NumPy")
plot_tensor(fig.add_subplot(gs[9, 1]), "Broadcasting", K_tf.numpy(), "TensorFlow")

# Flattening
L_np = A_np.flatten()
L_tf = tf.reshape(A_tf, (-1,))
ax_flat_np = fig.add_subplot(gs[10, 0])
ax_flat_np.bar(range(len(L_np)), L_np, color='blue')
ax_flat_np.set_title("Flatten (NumPy)", fontsize=10, pad=10)
ax_flat_tf = fig.add_subplot(gs[10, 1])
ax_flat_tf.bar(range(len(L_tf)), L_tf.numpy(), color='orange')
ax_flat_tf.set_title("Flatten (TensorFlow)", fontsize=10, pad=10)

# --- 2.3.4 Reduction Operations ---
# Sum
M_np = np.sum(A_np, axis=0)
M_tf = tf.reduce_sum(A_tf, axis=0)
ax_sum_np = fig.add_subplot(gs[11, 0])
ax_sum_np.bar(range(len(M_np)), M_np, color='blue')
ax_sum_np.set_title("Sum (NumPy)", fontsize=10, pad=10)
ax_sum_tf = fig.add_subplot(gs[11, 1])
ax_sum_tf.bar(range(len(M_tf)), M_tf.numpy(), color='orange')
ax_sum_tf.set_title("Sum (TensorFlow)", fontsize=10, pad=10)

# Mean
N_np = np.mean(A_np, axis=0)
N_tf = tf.reduce_mean(A_tf, axis=0)
ax_mean_np = fig.add_subplot(gs[12, 0])
ax_mean_np.bar(range(len(N_np)), N_np, color='blue')
ax_mean_np.set_title("Mean (NumPy)", fontsize=10, pad=10)
ax_mean_tf = fig.add_subplot(gs[12, 1])
ax_mean_tf.bar(range(len(N_tf)), N_tf.numpy(), color='orange')
ax_mean_tf.set_title("Mean (TensorFlow)", fontsize=10, pad=10)

# Argmax
O_np = np.argmax(A_np, axis=0)
O_tf = tf.argmax(A_tf, axis=0)
ax_argmax_np = fig.add_subplot(gs[13, 0])
ax_argmax_np.bar(range(len(O_np)), O_np, color='blue')
ax_argmax_np.set_title("Argmax (NumPy)", fontsize=10, pad=10)
ax_argmax_tf = fig.add_subplot(gs[13, 1])
ax_argmax_tf.bar(range(len(O_tf)), O_tf.numpy(), color='orange')
ax_argmax_tf.set_title("Argmax (TensorFlow)", fontsize=10, pad=10)

# Adjust layout and save
plt.tight_layout()
plt.savefig("tensor_operations_results.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Set random seed for reproducibility
tf.random.set_seed(42)

# Define tensors
A = tf.constant(np.random.rand(3, 3), dtype=tf.float32)
B = tf.constant(np.random.rand(3, 3), dtype=tf.float32)

# --- 2.3.1 Element-wise Operations ---
print("=== Element-wise Operations (TensorFlow) ===")
print("A:\n", A.numpy())
print("B:\n", B.numpy())
print("Addition (A + B):\n", (A + B).numpy())
print("Subtraction (A - B):\n", (A - B).numpy())
print("Hadamard Product (A * B):\n", (A * B).numpy())

# Activation Functions
relu = tf.nn.relu(A)
sigmoid = tf.nn.sigmoid(A)
print("ReLU(A):\n", relu.numpy())
print("Sigmoid(A):\n", sigmoid.numpy())

# Plot activation functions
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(relu.numpy(), cmap='viridis')
plt.colorbar()
plt.title("ReLU(A)")
plt.subplot(1, 2, 2)
plt.imshow(sigmoid.numpy(), cmap='viridis')
plt.colorbar()
plt.title("Sigmoid(A)")
plt.tight_layout()
plt.show()

# --- 2.3.2 Linear Algebra Operations ---
print("\n=== Linear Algebra Operations (TensorFlow) ===")
print("Matrix Multiplication (A @ B):\n", tf.matmul(A, B).numpy())
print("Transpose of A:\n", tf.transpose(A).numpy())

# Eigenvalues and Eigenvectors
eigvals, eigvecs = tf.linalg.eigh(A)  # For symmetric matrices
print("Eigenvalues of A:\n", eigvals.numpy())
print("Eigenvectors of A:\n", eigvecs.numpy())

# --- 2.3.3 Shape Manipulations ---
print("\n=== Shape Manipulations (TensorFlow) ===")
print("Reshaped A (9,1):\n", tf.reshape(A, (9, 1)).numpy())
print("Broadcasting (A + [1, 2, 3]):\n", (A + tf.constant([1, 2, 3], dtype=tf.float32)).numpy())
print("Flattened A:\n", tf.reshape(A, (-1,)).numpy())

# --- 2.3.4 Reduction Operations ---
print("\n=== Reduction Operations (TensorFlow) ===")
print("Sum of A (axis=0):\n", tf.reduce_sum(A, axis=0).numpy())
print("Mean of A (axis=0):\n", tf.reduce_mean(A, axis=0).numpy())
print("Argmax of A (axis=0):\n", tf.argmax(A, axis=0).numpy())
```
